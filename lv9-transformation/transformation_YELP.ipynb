```python
b_dict = {}

for i in range(len(b)):
    b[i]['reviews'] = []
    b_dict[b[i]['business_id']] = b[i]
```

```python
for i in range(len(r)):
    b_dict[r[i]['business_id']]['reviews'].append(r[i])
```

```python
def clean_name(name):
    clean_name = ''.join(c for c in name if c.isalpha())    
    clean_name = clean_name.lower()
    return clean_name
```

```python
b_dict_2 = {}

for k in list(b_dict.keys()):
    if b_dict[k]['reviews'] == []:
        pass
    
    else:
        b_dict[k]['inspections'] = []

        b_dict_2[clean_name(b_dict[k]['name'])] = b_dict[k]
```

```python
df.Name = df.Name.apply(clean_name)
```

```python
for k in list(b_dict_2.keys()):
    b_dict_2[k]['inspections'] = df[df.Name == k].to_dict('records')
```

```python
b_dict_2['yushan']
```

```python
import nltk

nltk.download('punkt')
nltk.download('stopwords')

stop_words = set(nltk.corpus.stopwords.words("english"))
```

```python
clean = []
dirty = []

for k in list(b_dict_2.keys()):
    
    try:
    
        if max(pd.DataFrame(b_dict_2[k]['inspections'])['Inspection Score']) < 20:
            clean.append(k)

        if max(pd.DataFrame(b_dict_2[k]['inspections'])['Inspection Score']) > 70:
            dirty.append(k)
    
    except:
        pass
```

```python
clean; dirty
```

```python
import re

regex = re.compile('[^a-zA-Z] ')

clean_words = []

for r in clean:
    text = pd.DataFrame(b_dict_2[r]['reviews'])['text']
    
    clean_words.append(sum(text.apply(lambda x: nltk.word_tokenize(regex.sub('', x.lower()))), []))
```

```python
dirty_words = []

for r in dirty:
    text = pd.DataFrame(b_dict_2[r]['reviews'])['text']
    
    dirty_words.append(sum(text.apply(lambda x: nltk.word_tokenize(regex.sub('', x.lower()))), []))
```

```python
clean_w_f = sum(clean_words, [])

dirty_w_f = sum(dirty_words, [])
```

```python
import collections
```

```python
collections.Counter(clean_w_f).most_common(25)
```

```python
collections.Counter(dirty_w_f).most_common(25)
```

```python
common = collections.Counter(clean_w_f).keys() & collections.Counter(dirty_w_f).keys()
common = list(common)
```

```python
result = pd.DataFrame()

for word in common:
    score_word = (collections.Counter(dirty_w_f)[word] / len(dirty_w_f))/(collections.Counter(clean_w_f)[word] / len(clean_w_f))
    
    dummy = pd.DataFrame({'word': word, 'score_word': score_word}, index=[])
    
    result = pd.concat([result, dummy])
```
