```python
r = [json.loads(line) for line in open('Seattle_Health_Code/review.json', 'r')]

r
```

```python
b_dict = {}

for i in range(len(b)):
    b[i]['reviews'] = []
    b_dict[b[i]['business_id']] = b[i]
```

```python
for i in range(len(r)):
    b_dict[r[i]['business_id']]['reviews'].append(r[i])
```

```python
def clean_name(name):
    clean_name = ''.join(c for c in name if c.isalpha())    
    clean_name = clean_name.lower()
    return clean_name
```

```python
b_dict_2 = {}

for k in list(b_dict.keys()):
    if b_dict[k]['reviews'] == []:
        pass
    
    else:
        b_dict[k]['inspections'] = []

        b_dict_2[clean_name(b_dict[k]['name'])] = b_dict[k]
```

```python
df.Name = df.Name.apply(clean_name)
```

```python
for k in list(b_dict_2.keys()):
    b_dict_2[k]['inspections'] = df[df.Name == k].to_dict('records')
```
